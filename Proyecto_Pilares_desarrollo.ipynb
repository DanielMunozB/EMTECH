Primero, diseñamos una función para obtener los promedios de cada pilar para todos los datos de la muestra. Esto, por medio de la suma de las conlumnas que pertenecen a los diferentes pilares (x). De esta manera, si deseamos obtener un promedio de un pilar en específico ya no será necesario computar la suma completa,sólo utilizalemos la funcion "promedioXpilares(x)".
